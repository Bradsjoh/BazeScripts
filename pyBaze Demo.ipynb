```python
from pybaze import pyBaze
```

```python
serverName = 'https://demo.bazefield.com/Bazefield.Services/api/'
user = 'Administrator'
t09wdspd = 'measurements/WFU-T09-WindSpeed/aggregates/3/from/*-7d/to/*/interval/600000?format=json'
t09pwr = 'measurements/WFU-T09-ActivePower/aggregates/3/from/*-7d/to/*/interval/600000?format=json'
t11wdspd = 'measurements/WFU-T11-WindSpeed/aggregates/3/from/*-7d/to/*/interval/600000?format=json'
t11pwr = 'measurements/WFU-T11-ActivePower/aggregates/3/from/*-7d/to/*/interval/600000?format=json'
```

```python
client = pyBaze(serverName, user)
```

```python
t09data = {'wind':client.getJsonData(t09wdspd), 'power':client.getJsonData(t09pwr)}
t11data = {'wind':client.getJsonData(t11wdspd), 'power':client.getJsonData(t11pwr)}
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
pcplot = plt.plot(t09data['wind']['v'], t09data['power']['v'],'bo', label = 'T09', markersize=3)
plt.plot(t11data['wind']['v'], t11data['power']['v'],'go', label = 'T11',  markersize=3)
plt.legend(loc=2)
```

```python
tplot = plt.plot(t09data['wind']['t'], t09data['wind']['v'],'bo', label = 'T09', markersize=3)
plt.plot(t11data['wind']['t'], t11data['wind']['v'],'go', label = 'T11',  markersize=3)
plt.legend(loc=4)
```
